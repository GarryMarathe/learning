```python
import tensorflow as tf
from tensorflow import keras
from keras.applications.vgg16 import VGG16
model = VGG16()
```

```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
```

```python

import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_folder = "C:\\Users\\Gaurav\\Desktop\\DL Lab Final\\images"
```

```python
image_files = [os.path.join(image_folder,f) for f in os.listdir(image_folder) if f.lower().endswith(('.jpg','.jpeg','.png', '.gif', '.bmp'))]
```

```python
image_data = []
for image_path in image_files:
    image = load_img(image_path,target_size=(224,224))
    image = img_to_array(image)
    image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
    image = preprocess_input(image)
    image_data.append(image)
```

```python
random_index = random.randint(0, len(image_data) - 1)
random_image = image_data[random_index]
```

```python
yhat = model.predict(random_image)
```

```python
label = decode_predictions(yhat)
```

```python
label = label[0][0]
```

```python
plt.title(label)
img = mpimg.imread(image_files[random_index])
plt.imshow(img)
plt.show()
```
